```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
salary_data = pd.read_csv('/content/drive/MyDrive/Python Projects/Machine Learning Projects/SalaryPrediciton/hiring.csv')
salary_data
```

```python
import math
test_score = salary_data['test_score(out of 10)']
test_score
```

```python
mean_test_score = math.floor(test_score.mean())
mean_test_score 
```

```python
salary_data['test_score(out of 10)'] = salary_data['test_score(out of 10)'].fillna(mean_test_score)
salary_data
```

```python
salary_data['experience'][2] = 5.0
salary_data['experience'][3] = 2.0
salary_data['experience'][4] = 7.0
salary_data['experience'][5] = 3.0
salary_data['experience'][6] = 10.0
salary_data['experience'][7] = 11.0
salary_data
```

```python
salary_data['experience'] = salary_data['experience'].fillna(0)
salary_data
```

```python
reg = linear_model.LinearRegression()
reg.fit(salary_data[['experience','test_score(out of 10)','interview_score(out of 10)']],salary_data['salary($)'])
```

Using our linear regression model, we can predict the salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

```python
reg.predict([[2,9,6],[12,10,10]])
```
